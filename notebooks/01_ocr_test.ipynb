```python
# 2. Import your agent and necessary tools
from src.agents.vision import extract_prescription_data
from IPython.display import Image, display
import json
import glob
```

```python
# 3. Load your Kaggle Images
# (Assuming you unzipped the dataset into data/medical_images)
image_paths = glob.glob("../data/medical_images/*.jpg")

# Check if we found images
print(f"Found {len(image_paths)} images.")
```

```python

# 4. The Test Loop (Run this cell to test)
# Let's test on the first 3 images
for img_path in image_paths[:3]:
    print(f"\n--- Processing: {os.path.basename(img_path)} ---")
    
    # A. Display the image so YOU can read it
    display(Image(filename=img_path, width=400))
    
    # B. Run your Vision Agent
    try:
        result = extract_prescription_data(img_path)
        
        # C. Print the Structured JSON
        # (Using .dict() if it's Pydantic v1 or .model_dump() for v2)
        print("✅ Extracted Data:")
        print(json.dumps(result.dict(), indent=2)) 
        
    except Exception as e:
        print(f"❌ Error: {e}")
```
